```python
# Import necessary library
import pandas as pd
import numpy as np
```

```python
# Load your CSV (correcting encoding)
df = pd.read_csv("ycombinator_company_not_clean_data.csv", encoding="latin1")
df.head(10)  # Display the first 10 rows of the DataFrame
```

```python
df[['founder_1', 'founder_2', 'founder_3']]
# These columns contain founder information, which will need cleaning.
```

```python
# Function to clean founder info
def clean_founder(text):
    if pd.isna(text) or text == "Not available":
        return "Not available"
    # Take only the first line (because 1st line has the name)
    return text.split('\n')[0].strip()

# Apply to founder columns
for i in range(1, 4):
    df[f'founder_{i}'] = df[f'founder_{i}'].apply(clean_founder)
```

```python
df[['founder_1', 'founder_2', 'founder_3']]
```

```python
df.count()  # Count non-null values in each column
```

```python
df.info()  # Display DataFrame info to check data types and non-null counts
```

```python
df['company_name'].duplicated().any()
# Check for duplicates in a specific column
```

```python
# Show all duplicate values in the column
duplicates = df[df['company_name'].duplicated(keep=False)]['company_name']
print(duplicates)
```

```python
# Count occurrences of each company name

duplicate_counts = df['company_name'].value_counts()
duplicates_only = duplicate_counts[duplicate_counts > 1]
print(duplicates_only)
```

```python
# Remove duplicates, keeping the first occurrence

df.replace("Not available", pd.NA, inplace=True) # Replace 'Not available' with NaN

# Count non-NA values per row
df['non_missing_count'] = df.notna().sum(axis=1)

# Sort so rows with more data come first, then drop duplicates keeping the first
df_cleaned = df.sort_values('non_missing_count', ascending=False)\
               .drop_duplicates(subset='company_name', keep='first')\
               .drop(columns='non_missing_count')

# Replace NaN back with 'Not available'
df_cleaned.fillna("Not available", inplace=True)
```

```python
# check for duplicates in the cleaned DataFrame
df_cleaned['company_name'].duplicated().any()
```

```python
df_cleaned.count()  # after cleaning Count non-null values in each column
```

```python
df_cleaned['founded_date'].isna()  # Check the distribution of founded dates
```

```python
# Replace "Not available" with NaN
df_cleaned['founded_date'] = df_cleaned['founded_date'].replace("Not available", np.nan)
```

```python
# Optional: convert to numeric (errors='coerce' converts non-numeric to NaN)
df['founded_date'] = pd.to_numeric(df['founded_date'], errors='coerce')
```

```python
df_cleaned['founded_date'].value_counts()  # Check the distribution of founded dates
```

```python
# Save cleaned CSV

df_cleaned.to_csv("yc_company_data.csv", index=False, encoding="utf-8-sig")
print("Cleaned data saved successfully.")
```
